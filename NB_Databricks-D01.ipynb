Day-01: Databricks Challenge

```python
from pyspark.sql import functions as F
```

```python
# Create simple DataFrame
data = [("iPhone", 999), ("Samsung", 799), ("MacBook", 1299)]
df = spark.createDataFrame(data, ["product", "price"])
df.show()

# Filter expensive products
df.filter(df.price > 1000).show()
```

```python
# Add a 10% tax column and a category column
df_transformed = df.withColumn("tax", df.price * 0.10) \
                   .withColumn("is_premium", F.when(df.price > 1000, "Yes").otherwise("No"))
df_transformed.show()
```

```python
# Group by product and to find average price
df.groupBy("product").agg(F.avg("price").alias("avg_price")).show()
```

```python
# Convert product names to uppercase and get the first 3 letters
df.withColumn("upper_product", F.upper(F.col("product"))) \
  .withColumn("code", F.substring(F.col("upper_product"), 1, 3)) \
  .show()
```
