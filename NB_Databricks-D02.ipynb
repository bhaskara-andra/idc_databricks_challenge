Day-02: Databricks Challenge

```python
from pyspark.sql.functions import col, month, year, max as spark_max, count, to_date, to_timestamp, month, year, lit, countDistinct
```

```python
file_path = f"/databricks-datasets/online_retail/data-001/*"
```

```python
try:
    df = spark.read.csv(file_path, header=True, inferSchema=True)
    display(df.limit(10))
except Exception as e:
    print(f"Error reading CSV file: {e}")
```

```python
df.printSchema()
```

```python
morph_dtype_df = df.withColumn("InvoiceDate", to_timestamp("InvoiceDate", "M/d/yy H:mm"))
# df.select(to_date(col("InvoiceDate")).alias("InvoiceDate"))
```

```python
max_date = morph_dtype_df.select(spark_max("InvoiceDate")).collect()[0][0]
last_month = max_date.month
last_year = max_date.year
```

```python
result_df = morph_dtype_df.filter((month("InvoiceDate") == last_month) & 
                         (year("InvoiceDate") == last_year)) \
    .groupBy("CustomerID") \
    .agg(countDistinct("InvoiceNo").alias("VisitCount")) \
    .orderBy(col("VisitCount").desc())
```

```python
display(result_df)
```

```python

# Filter for the latest month and group by CustomerID to count orders
orders_last_month = (
    df.filter((month(col("InvoiceDate")) == latest_month) & (year(col("InvoiceDate")) == latest_year))
      .groupBy(col("CustomerID"))
      .agg(count(col("InvoiceNo")).alias("num_orders"))
)

display(orders_last_month)
```
